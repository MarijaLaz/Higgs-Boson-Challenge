## Load the training data into feature matrix, class labels, and event ids:

```python
#load data
DATA_TRAIN_PATH = '../data/train.csv' #download train data and supply path here 
y, x, ids = load_csv_data(DATA_TRAIN_PATH)

#add constant term
#tx = np.c_[np.ones((y.shape[0], 1)), x]

print(y.shape, x.shape, ids.shape)
```

```python
jet_0_tx, jet_1_tx, jet_2_tx, jet_3_tx, jet_0_y, jet_1_y, jet_2_y, jet_3_y, index_0, index_1, index_2, index_3 = split_jet(y, x)
```

```python
replaceNaN(jet_2_tx)
jet_2_tx = standardize(jet_2_tx)
```

```python
#without feature extension
acc_LR = []
total_loss_te_LR = []
acc_RLR = []
total_loss_te_RLR = []

gamma = 1e-6
max_iters = 100
lambda_ = 0.001
k_fold = 5

initial_w = np.zeros(jet_2_tx.shape[1])

k_indices = build_k_indices(jet_2_y, k_fold, 1)


for k in range(k_fold):
    acc, loss_te_LR = cross_validation(jet_2_y, jet_2_tx, k_indices, k, initial_w, 'logistic_regression', max_iters, gamma, lambda_)
    acc_LR.append(acc)
    total_loss_te_LR.append(loss_te_LR)

print(acc_LR)
print(total_loss_te_LR)

```

```python
#with extension
acc_LR = []
total_loss_te_LR = []
acc_RLR = []
total_loss_te_RLR = []

gamma = 1e-6
max_iters = 100
lambda_ = 0.001
k_fold = 5
degree=3

jet_2_tx_poly=buil_poly(jet_2_tx,degree)
initial_w = np.zeros(jet_2_tx_poly.shape[1])

k_indices = build_k_indices(jet_2_y, k_fold, 1)


for k in range(k_fold):
    acc, loss_te_LR = cross_validation(jet_2_y, jet_2_tx_poly, k_indices, k, initial_w, 'logistic_regression', max_iters, gamma, lambda_)
    acc_LR.append(acc)
    total_loss_te_LR.append(loss_te_LR)

print(acc_LR)
print(total_loss_te_LR)
```
